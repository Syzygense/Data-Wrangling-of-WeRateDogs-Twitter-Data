## Data Wrangling of Twitter WeRateDogs Data

### Introduction

WeRateDogs is a Twitter account that rates people's dogs with humorous comments about the dog. In this project, we perform data analysis on the WeRateDogs Twitter account with a focus on the data wrangling process. Basic data visualization is also performed to display the results.

```python
# Import necessary packages
import pandas as pd
import tweepy
import json
import time
```

### Gather

Data is provided from the WeRateDogs Twitter archive in the 'twitter-archive-enhanced.csv' file. This file contains data about the dog's rating, name and stage.

```python
# Read 'twitter-archive-enhanced.csv' file into a pandas DataFrame
archive_df = pd.read_csv('twitter-archive-enhanced.csv')
```

Additional data is provided through a neural network that can classify the breeds of dogs. The result is a table of image predictions in the 'image-predictions.tsv' file. 

```python
# Read 'image-predictions.tsv' file into a pandas DataFrame
image_df = pd.read_csv('image-predictions.tsv', sep='\t')
```

We further obtain data from the Twitter API, which includes valuable and interesting information such as retweet count and favorite count, among other useful data. The data is obtained by querying the Twitter API for each Tweet's JSON data using Python's Tweepy library.

```python
# Create an API object to gather Twitter data
consumer_key = 'Confidential - Not shown'
consumer_secret = 'Confidential - Not shown'
access_token = 'Confidential - Not shown'
access_secret = 'Confidential - Not shown'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
# Gather Twitter data as a JSON object using the Tweet IDs and write line-by line
# into the 'tweet_json.txt' file
start_tot = time.time()
error_count = 0
with open('tweet_json.txt', 'w') as f:
    for id_num in archive_df.tweet_id:
        start = time.time()
        try:
            tweet = api.get_status(id_num, tweet_mode='extended')
            json.dump(tweet._json, f)
            f.write('\n')
            print('Tweet ID: ' + str(id_num))
        except:
            print('Error in retrieving tweet')
            error_count += 1
        end = time.time()
        print('Loop runtime: ' + str(end - start) + ' s')
print('The number of errors is ' + str(error_count))
end_tot = time.time()
print('The total runtime is ' + str(end_tot - start_tot) + ' s')
```

```python
# Read JSON data from the 'tweet_json.txt' file into a pandas DataFrame
twitter_dict = {'tweet_id': [], 'retweet_count': [], 'likes_count': []}

with open('tweet_json.txt') as f:
    tweet_str = f.readline()
    while tweet_str:
        tweet = json.loads(tweet_str)
        twitter_dict['tweet_id'].append(tweet['id'])
        twitter_dict['retweet_count'].append(tweet['retweet_count'])
        twitter_dict['likes_count'].append(tweet['favorite_count'])
        tweet_str = f.readline()

api_df = pd.DataFrame(twitter_dict)
```

### Clean

We will programmatically clean the data, starting from the quality issues.

```python
# Copy pandas dataframes
archive_df_clean = archive_df.copy()
image_df_clean = image_df.copy()
api_df_clean = api_df.copy()
```

##### Define

- Remove retweets and replies by using df.drop() to drop rows where the values in 'retweeted_status' and 'in_reply_to_status' are not null

##### Code

```python
archive_df_clean.drop(archive_df_clean[archive_df_clean.in_reply_to_status_id.notnull()].index, inplace=True)
archive_df_clean.drop(archive_df_clean[archive_df_clean.retweeted_status_id.notnull()].index, inplace=True)
```

##### Test

```python
# See if there are any not null values
archive_df_clean[archive_df_clean.in_reply_to_status_id.notnull()]
```

```python
archive_df_clean[archive_df_clean.retweeted_status_id.notnull()]
```

##### Define

- Drop the rows with missing expanded_urls

##### Code

```python
archive_df_clean.drop(archive_df_clean[archive_df_clean.expanded_urls.isnull()].index, inplace=True)
```

##### Test

```python
# There should be no data missing now from null values
archive_df_clean.info()
```

##### Define

- Find the invalid dog names by looking whether the first letter is capitalized and replace them with 'None'

##### Code

```python
# Write loop to replace invalid dog names with 'None'. Documentation for how to
# loop through pandas dataframes was found from 
# https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
# https://stackoverflow.com/questions/23330654/update-a-dataframe-in-pandas-while-iterating-row-by-row
for index, row in archive_df_clean.iterrows():
    if row['name'][0].islower():
        archive_df_clean.set_value(index, 'name', 'None')
```

##### Test

```python
# Loop through dog names and print lower case names
for index, row in archive_df_clean.iterrows():
    if row['name'][0].islower():
        print(row['name'])
```

##### Define

- Change the datatype of the 'timestamp' column to datetime using pd.to_datetime() (documentation found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html)

##### Code

```python
archive_df_clean.timestamp = pd.to_datetime(archive_df_clean.timestamp)
```

##### Test

```python
# Check the datatype of the 'timestamp' column
archive_df_clean.timestamp.dtype
```

##### Define

- Convert 'rating_numerator' to type float and extract the proper ratings from the text

##### Code

```python
# Convert column to type float
archive_df_clean.rating_numerator = archive_df_clean.rating_numerator.astype('float64')

# Extract numerator ratings with decimals from the text
decimal_ratings = archive_df_clean[archive_df_clean.text.str.contains(r"(\d+\.\d*\/\d+)")].text.str.extract(r"(\d+\.\d*\d+)", expand=True).astype('float64')
#archive_df_clean[archive_df_clean.text.str.contains(r"(\d+\.\d*\/\d+)")]['rating_numerator']
for index, row in decimal_ratings.iterrows():
    archive_df_clean.set_value(index, 'rating_numerator', row[0])
```

##### Test

```python
# Check that the type has become float
archive_df_clean.rating_numerator.dtype
```

```python
# Check that numerator ratings with decimals have been properly extracted to column
archive_df_clean[archive_df_clean.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_numerator']]
```

##### Define

- For the image predictions, replace the underscore word separators with spaces and capitalize first letter using str.replace() (documentation found at https://www.tutorialspoint.com/python/string_replace.htm) and str.capitalize (documentation found at https://www.geeksforgeeks.org/string-capitalize-python/)

##### Code

```python
# Replace underscores with spaces
image_df_clean.p1 = image_df_clean.p1.str.replace('_', ' ')
image_df_clean.p2 = image_df_clean.p2.str.replace('_', ' ')
image_df_clean.p3 = image_df_clean.p3.str.replace('_', ' ')

# Capitalize first letter
image_df_clean.p1 = image_df_clean.p1.str.capitalize()
image_df_clean.p2 = image_df_clean.p2.str.capitalize()
image_df_clean.p3 = image_df_clean.p3.str.capitalize()
```

```python
# Find underscores to see if they are still there
print(sum(image_df_clean.p1.str.find('_') != -1))
print(sum(image_df_clean.p2.str.find('_') != -1))
print(sum(image_df_clean.p3.str.find('_') != -1))
```

```python
# Check if first letter is capitalized
print(sum(image_df_clean.p1.str[0].str.islower()))
print(sum(image_df_clean.p2.str[0].str.islower()))
print(sum(image_df_clean.p3.str[0].str.islower()))
```

##### Define

- Create columns with descriptive names for best image prediction of dog breed

##### Code

```python
# Write loop to create column from first prediction where dog breed is True
for index, row in image_df_clean.iterrows():
    if row['p1_dog'] == True:
        image_df_clean.set_value(index, 'dog_breed', row['p1'])
        image_df_clean.set_value(index, 'dog_breed_confidence', row['p1_conf'])
    elif row['p2_dog'] == True:
        image_df_clean.set_value(index, 'dog_breed', row['p2'])
        image_df_clean.set_value(index, 'dog_breed_confidence', row['p2_conf'])
    elif row['p3_dog'] == True:
        image_df_clean.set_value(index, 'dog_breed', row['p3'])
        image_df_clean.set_value(index, 'dog_breed_confidence', row['p3_conf'])
    else:
        image_df_clean.set_value(index, 'dog_breed', 'No dog breed found')
        image_df_clean.set_value(index, 'dog_breed_confidence', 0)
```

```python
# Visually inspect new 'dog_breed' and 'dog_breed_confidence' columns
image_df_clean
```

##### Define

- Make 'dog_stage' column and populate values with 'doggo', 'floofer', 'pupper' and 'puppo' data

##### Code

```python
# Write loop to populate 'dog_stage' column
for index, row in archive_df_clean.iterrows():
    stage = []
    if row['doggo'] != 'None':
        stage.append(row['doggo'])
    if row['floofer'] != 'None':
        stage.append(row['floofer'])
    if row['pupper'] != 'None':
        stage.append(row['pupper'])
    if row['puppo'] != 'None':
        stage.append(row['puppo'])
    if stage == []:
        stage.append('None')
    archive_df_clean.set_value(index, 'dog_stage', ', '.join(stage))
    
# Drop previous dog stage columns as no longer needed
archive_df_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

##### Define

- Merge dog breed image prediction data and API data with retweet count and favorite count into master dataset
- Use an inner join with the Tweet ID to reconciliate missing rows

##### Code

```python
# Merge 'image_df_clean' table with an inner join
# Documentation was found at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
twitter_archive_master_df = archive_df_clean.copy()
twitter_archive_master_df = twitter_archive_master_df.merge(image_df_clean[['tweet_id', 'jpg_url', 'img_num', 'dog_breed', 'dog_breed_confidence']], on='tweet_id')
```

### Analysis and Visualization

Now that the data wrangling is complete, we can analyze and visualize the data. We will investigate the data by plotting charts for retweets and likes count for the dog stages. Also, we will display the relationship between the rating and certain dog breeds.

```python
# Plot bar chart for average retweets count for each dog stage
% matplotlib inline
import matplotlib.pyplot as plt

twitter_archive_master_df.groupby(['dog_stage']).retweet_count.mean().plot.bar(color='gray');
plt.ylabel('Average Retweets');
plt.xlabel('Dog Stage');
plt.title('Average Retweets of Different Dog Stages');
```

```python
# Plot bar chart for average likes count for each dog stage
twitter_archive_master_df.groupby(['dog_stage']).likes_count.mean().plot.bar(color='gray');
plt.ylabel('Average Likes');
plt.xlabel('Dog Stage');
plt.title('Average Likes of Different Dog Stages');
```

```python
# For the ratings analysis and visualization, we first calculate the rating 
# as a decimal
twitter_archive_master_df['rating_decimal'] = twitter_archive_master_df['rating_numerator'] / twitter_archive_master_df['rating_denominator']
```

```python
# Plot bar charts for the ratings of the 5 most common dog breeds
twitter_archive_master_df.dog_breed.value_counts() 

dog_breed_labels = ['Golden retriever', 'Labrador retriever', 'Pembroke', 'Chihuahua', 'Pug']
dog_breed_mask = twitter_archive_master_df['dog_breed'].isin(dog_breed_labels)
twitter_archive_master_df[dog_breed_mask].groupby('dog_breed').rating_decimal.mean().plot.bar(color='gray');
plt.ylabel('Average Normalized Rating');
plt.xlabel('Dog Breed');
plt.title('Average Normalized Ratings of 5 Most Common Dog Breeds');
```

Retweets and likes counts can give an indication of the popularity of the Tweet. From the retweets and likes counts at various dog stages, we see that the retweets and likes counts follow similar trends, so dog stages with higher retweets counts also had higher likes counts. Dogs that were doggos and puppos were the more popular dog stages compared to floofers and puppers. Additionally, dogs that were both doggos and puppos had the highest retweets and likes counts. Tweets where a dog stage was mentioned were generally more popular than if a dog stage was not mentioned (value is 'None'), with the exception of pupper.

For the 5 most common dog breeds, we see that the normalized average ratings expressed as a decimal are similar, and are on average larger than 1. This suggests that the ratings given by WeRateDogs are consistent across different dog breeds. The 5 most common dog breeds were Chihuahua, Golden Retriever, Labrador Retriever, Pembroke and Pug.
